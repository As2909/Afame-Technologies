1. Eliminate the NaN values

```python
# calcualte the total no. of null values in every columns
hr_df.isnull().sum()

# There is zero null values
```

2. Drop the unnecessay Columns

```python
# remove the unnecessary columns
columns_to_drop = ["DailyRate", "BusinessTravel", "DistanceFromHome","EmployeeCount",
                   "EnvironmentSatisfaction", "HourlyRate", "JobSatisfaction",
                   "Over18"]
# Create new Dataframe to store the cleaned dataframe
hr_df_cleaned = hr_df.drop(columns_to_drop,axis=1)
```

3. Giving the columns new names

```python
new_columns_names = {
    "Department": "Dept",
    "Education": "EduLevel",
    "EmployeeNumber": "EmployeeId",
    "MaritalStatus": "Marital",
    "MonthlyIncome": "Income"
}

hr_df_cleaned = hr_df_cleaned.rename(columns=new_columns_names, )
```

```python
# Rename the row_labels

# rename In Dept column
hr_df_cleaned.loc[hr_df_cleaned["Dept"] == "Research & Development", "Dept"] = "R&D"

#rename in Gender coloumn
hr_df_cleaned.loc[hr_df_cleaned["Gender"] == "Female", "Gender"] = "F"
hr_df_cleaned.loc[hr_df_cleaned["Gender"] == "Male", "Gender"] = "M"
```
