2. Analyze sales trends over Time

Monthly Sales Over Time

```python
# cast the "Order Date" column into date format
sales_df["Order Date"] = pd.to_datetime(sales_df["Order Date"])
```

```python
# Set the "Order Date" Column As the index
sales_df.set_index("Order Date", inplace=True)
```

```python
# Group the data by month and sum the sales for each month of that year
monthly_sales = sales_df["Sales"].resample("M").sum()
```

```python
print(monthly_sales)
```

Monthly Sales Trends in over Time

```python
plt.figure(figsize=(10,6))
plt.plot(monthly_sales.index, monthly_sales, linewidth=1)
plt.title("Monthly Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

Yearly Sales

```python
# Group the data by year and sum the sales for each year
yearly_sales = sales_df["Sales"].resample("Y").sum()
```

```python
print(round(yearly_sales))
```

Yearly Sales Trends Over Time

```python
plt.figure(figsize=(10, 6))
yearly_sales.plot()
plt.title('Yearly Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

3.Best Selling Products

```python
# Grouped the data by "Prodcut Name" column
sales_df.groupby("Product Name")["Sales"].sum()
```

```python
product_df = sales_df.groupby("Product Name")["Sales"].sum()
```

```python
# Total Number of Unique Products
product_df.count()
```

```python
product_df.sort_values(ascending=False)
```

```python
sorted_product_df = product_df.sort_values(ascending=False)
```

Best Selling Products Top 5

```python
# Top 5 selling products
sorted_product_df.head()
```

```python
top_five_products_df = sorted_product_df.head()
```

```python
plt.figure(figsize=(10,6))
top_five_products_df.plot(kind="bar")
plt.title("Top 5 Product Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.show()
```
